```python
# ===== 00 Cleanup: remove old outputs safely =====
import shutil
from pathlib import Path

# 确定项目根目录
CWD = Path.cwd()
if (CWD / "outputs").exists():
    ROOT = CWD
elif (CWD.parent / "outputs").exists():
    ROOT = CWD.parent
else:
    raise RuntimeError(f"找不到 outputs 目录，请在项目根目录或 notebooks 目录运行。cwd={CWD}")

OUT = ROOT / "outputs"
print("Project root:", ROOT)
print("Outputs dir :", OUT)
```

```python
# ===== Preview files to be deleted =====

all_items = list(OUT.glob("*"))

if not all_items:
    print("outputs 目录已经是空的，无需清理。")
else:
    print("将删除以下内容：")
    for p in all_items:
        if p.is_dir():
            print(f"[DIR ] {p.relative_to(ROOT)}")
        else:
            print(f"[FILE] {p.relative_to(ROOT)}")
```

```python
# ===== Delete all contents under outputs =====

for p in OUT.iterdir():
    if p.is_dir():
        shutil.rmtree(p)
        print(f"Deleted directory: {p.name}")
    else:
        p.unlink()
        print(f"Deleted file     : {p.name}")

print("\n✅ outputs 目录已清空")
```

```python
# ===== Recreate standard output structure (optional) =====

subdirs = [
    "figures",
    "corr_advanced",
    "shap",
    "maps",
    "holdout_70_30",
    "qc"
]

for d in subdirs:
    (OUT / d).mkdir(parents=True, exist_ok=True)
    print(f"Created: outputs/{d}")

print("\n✅ 标准输出目录已重建")
```
